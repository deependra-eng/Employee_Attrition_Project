```python
#Label Encodeing
# Target Variable(Attrition)
data['Attrition'] = data['Attrition'].replace({'No':0,'Yes':1})
```

```python
#encode binary variables
data['OverTime'] = data['OverTime'].map({'No':0,'Yes':1})
data['Gender'] = data['Gender'].map({'Male':0,'Female':1})
```

```python
# encode categorical columns which are ordinal, use labelEncoding
# apply Label encoder to df_categorical
from sklearn.preprocessing import LabelEncoder
encoding_cols=['BusinessTravel','Department','EducationField','JobRole','MaritalStatus']
label_encoders = {}
for column in encoding_cols:
    label_encoders[column] = LabelEncoder()
    data[column] = label_encoders[column].fit_transform(data[column])
```

```python
# look at the final data
data.head()
```

```python
#Machine Learning: Splitting the data into Independent(X) and Dependent(y)
X = data.drop(['Attrition','Over18'], axis=1)
y = data['Attrition'].values
```

```python
# Resampling of Y variable
from collections import Counter
from imblearn.over_sampling import RandomOverSampler
print(Counter(y))
rus = RandomOverSampler(random_state = 42)
X_over, y_over = rus.fit_resample(X,y)
print(Counter(y_over))
```

```python
# Split the data into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_over, y_over, test_size=0.2, random_state=42)
```

```python
# Sanity check for the sampled data
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#Model-1 [Logistic Regression]
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
prediction1=logreg.predict(X_test)
print("Accuracy Score -", accuracy_score(y_test , prediction1))
```

```python
#Model-2 [Support Vector Classifier]
from sklearn.svm import SVC
sv = SVC()
sv.fit(X_train, y_train)
prediction2=sv.predict(X_test)
print("Accuracy Score -", accuracy_score(y_test , prediction2))
```

```python
#Model-3 [Decision Tree Classifier]
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
prediction3=dt.predict(X_test)
print("Accuracy Score -", accuracy_score(y_test , prediction3))
```

```python
#Model-4 [Random Forest Classifier]
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
prediction4=rf.predict(X_test)
print("Accuracy Score -", accuracy_score(y_test , prediction4))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
y_train_prob = rf.predict_proba(X_train)[:, 1]
y_test_prob = rf.predict_proba(X_test)[:, 1]

# Calculate ROC-AUC scores
train_auc = roc_auc_score(y_train, y_train_prob)
test_auc = roc_auc_score(y_test, y_test_prob)

print(f"Train ROC-AUC: {train_auc}")
print(f"Test ROC-AUC: {test_auc}")
```

```python
# Calculate ROC curve
fpr_train, tpr_train, _ = roc_curve(y_train, y_train_prob)
fpr_test, tpr_test, _ = roc_curve(y_test, y_test_prob)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr_train, tpr_train, label='Train ROC curve (area = %0.2f)' % train_auc)
plt.plot(fpr_test, tpr_test, label='Test ROC curve (area = %0.2f)' % test_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```
